```python
def categorize_score(score):
    """
    Categorizes a score into one of five categories based on the given thresholds.

    Parameters:
    score (int): The score to be categorized.

    Returns:
    str: The category of the score. Possible values are "Poor", "Low", "Medium", "Good", or "Excellent".
    """
    if score <= 20:
        return "Poor"
    elif score <= 40:
        return "Low"
    elif score <= 60:
        return "Medium"
    elif score <= 80:
        return "Good"
    else:
        return "Excellent"
```
