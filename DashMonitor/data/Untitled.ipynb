```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
df=pd.read_csv("Sentiment_Data_with_Normalized_Score.csv")
df.columns
```

```python
df['New_Sentiment_Score'].value_counts()
```

```python
df['data_source'] = df['data_source'].replace('maps', 'Google')

# Tomar el 25% de los datos y asignarles valores aleatorios
num_samples = int(0.25 * len(df))
random_indices = np.random.choice(df.index, num_samples, replace=False)

# Valores aleatorios para asignar
new_sources = np.random.choice(['Facebook', 'Yelp', 'News'], num_samples, replace=True)

# Asignar los nuevos valores a las filas seleccionadas
df.loc[random_indices, 'data_source'] = new_sources
```

```python
df['data_source'].value_counts()
```

```python
del df['Unnamed: 0']
df
```

```python
df['New_Sentiment_Score'].value_counts()
```

```python
df['New_Sentiment_Score'] = df['New_Sentiment_Score'].replace('Poor', 'Very Low')
df['New_Sentiment_Score'] = df['New_Sentiment_Score'].replace('Below Average', 'Low')
df['New_Sentiment_Score'] = df['New_Sentiment_Score'].replace('Average', 'Medium')
df['New_Sentiment_Score'] = df['New_Sentiment_Score'].replace('Good', 'High')
df['New_Sentiment_Score'] = df['New_Sentiment_Score'].replace('Excellent', 'Very High')
```

```python
def classify_score_new(score):
    if score == 'Poor':
        return 'Very Low'
    elif score == 'Below Average':
        return 'Low'
    elif score == 'Average':
        return 'Medium'
    elif score == 'Good':
        return 'High'
    else:
        return 'Very High'
        
```

```python
df.to_csv("Sentiment_Data_with_Normalized_Score.csv",index=False)
```

```python
df["ds"] = pd.to_datetime(df["ds"])
df1 = df.copy()
df["Nuevo_Sentiment_Score"] = df["Normalized_Sentiment_Score"]                

month_map = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}
df["month_num"] = df["month"].map(month_map)
df = (
    df.groupby(["Bank Name", "Predicted_Pilar", "year", "month_num"])[
        ["Nuevo_Sentiment_Score"]
    ]
    .mean()
    .reset_index()
)
df["date"] = pd.to_datetime(
    df["year"].astype(str) + "-" + df["month_num"].map("{:02}".format), format="%Y-%m"
)
df["Total_Sentiment_Score"] = df.groupby(["Bank Name", "date"])[
    "Nuevo_Sentiment_Score"
].transform("mean")
df.sort_values("date", inplace=True)
df
```
