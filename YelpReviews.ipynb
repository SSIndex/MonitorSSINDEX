```python
# Función para obtener y guardar las reseñas de Yelp de un banco
def get_and_save_yelp_reviews(val, bank, api_key):
    # Extraer el nombre del banco de la URL para nombrar el archivo CSV
    bank_name = bank.split("?")[0].split("/")[-1]
   
    # URL de la API de Outscraper
    url = f"https://api.app.outscraper.com/yelp/reviews?query={bank}&async=false"
    
    # Encabezados para la solicitud incluyendo la clave de la API
    headers = {"X-API-KEY": api_key}
    
    # Realizar la solicitud GET a la API
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        reviews_data = response.json().get('data', [])
        
        # Verificar que hay datos de reseñas
        if reviews_data:
            # La API devuelve una lista de listas de reseñas, por lo que se toma el primer elemento
            reviews_list = reviews_data[0]
            
            # Crear DataFrame a partir de los datos de las reseñas
            df_reviews = pd.DataFrame(reviews_list)
            
            # Guardar los datos en un archivo CSV
            filename = f"Yelp/{bank_name}_{val}.csv"
            df_reviews.to_csv(filename, index=False)
            print(f"Reseñas de '{bank_name}' guardadas correctamente.")
        else:
            print(f"No se encontraron reseñas para el banco: {bank_name}")
    else:
        print(f"Error en la petición para el banco {bank_name}: {response.status_code}")
```
